```python
import pandas as pd
df = pd.read_csv("KaggleV2-May-2016.csv.zip")
df.head()
```

```python
# Standardize column names: lowercase and replace hyphens with underscores
df.columns = [col.strip().lower().replace('-', '_') for col in df.columns]

# Show updated column names
df.columns
```

```python
# Convert the date columns to datetime
df['scheduledday'] = pd.to_datetime(df['scheduledday'])
df['appointmentday'] = pd.to_datetime(df['appointmentday'])

# Confirm the changes
df[['scheduledday', 'appointmentday']].head()
```

```python
# Calculate number of days between scheduling and appointment
df['waiting_days'] = (df['appointmentday'] - df['scheduledday']).dt.days

# Preview the result
df[['scheduledday', 'appointmentday', 'waiting_days']].head()
```

```python
# Filter out invalid rows
df = df[(df['age'] >= 0) & (df['waiting_days'] >= 0)]

# Confirm changes
df[['age', 'waiting_days']].describe()
```

```python
# Convert no_show column to numeric: Yes = 1, No = 0
df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})

# Check the result
df['no_show'].value_counts()
```

```python

df['age_group'] = pd.cut(df['age'], bins=[0, 18, 40, 60, 100], labels=['0–18', '19–40', '41–60', '61+'])
df['weekday'] = df['appointmentday'].dt.day_name()

# Save to CSV
df.to_csv("cleaned_appointments_for_powerbi.csv", index=False)

# Download the file
from google.colab import files
files.download("cleaned_appointments_for_powerbi.csv")
```
